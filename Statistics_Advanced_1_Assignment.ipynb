Question 1: What is a random variable in probability theory?

Answer: A random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment. It maps outcomes of a random process to numbers, allowing us to apply mathematical analysis and compute probabilities, expectations, variances, etc.

Question 2: What are the types of random variables?

Answer: There are two types of random variables:

* Discrete random variable: A discrete random variable takes on a finite or countably infinite set of values.
* Continuous random variable: A continuous random variable takes on uncountably infinite values within a given interval or on the real number line.

Question 4. What is a binomial distribution, and how is it used in probability?

Answer: The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes: success or failure.

The binomial distribution is used when you want to model the number of times something happens in a fixed number of trials, such as:
* Number of defective items in a batch
* Number of correct answers on a multiple-choice test
* Number of people who respond "yes" in a survey
* Number of customers making a purchase

Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer: The Central Limit Theorem (CLT) is one of the most fundamental theorems in statistics. It explains why and how we can use normal distributions to make inferences about populations, even when the population itself is not normally distributed.

It is critical because:

‚úÖ 1. Enables use of the normal distribution:
* It allows us to use normal probability techniques to analyze sample means, even when the population is not normal.

‚úÖ 2. Foundation for confidence intervals and hypothesis testing:
* Most statistical methods (confidence intervals, t-tests, z-tests, etc.) rely on the assumption that sample means are approximately normal. The CLT justifies this.

‚úÖ 3. Works even if the original data isn't normal:
* For large enough n, the shape of the population distribution doesn't matter (as long as it has a finite mean and variance).

‚úÖ 4. Improves with larger samples:
* The larger the sample size n, the closer the sample mean distribution is to a true normal distribution.

Question 7: What is the significance of confidence intervals in statistical analysis?

Answer: A confidence interval is a range of values, derived from a sample, that is likely to contain the true population parameter with a certain level of confidence.

Significance:

‚úÖ 1. They quantify uncertainty:
* A point estimate (like a sample mean) tells you a single value ‚Äî but it doesn't tell you how reliable that estimate is.
* A confidence interval expresses both the estimate and the margin of error.

‚úÖ 2. They allow for better decision-making:
* Policymakers, scientists, and analysts use CIs to evaluate whether results are statistically meaningful or practically important.
* Example: If a 95% CI for a drug‚Äôs effect is (1.2, 3.8), and it doesn‚Äôt include 0, the drug is likely effective.

‚úÖ 3. They reflect sample variability:
* Different samples give different estimates. CIs account for this variability and reflect how much the estimate might vary due to random sampling.

‚úÖ 4. They are foundational in hypothesis testing:
* Confidence intervals and hypothesis tests are mathematically linked.
* If a 95% CI for a difference excludes 0, that corresponds to rejecting the null hypothesis at the 5% level.

‚úÖ 5. They help avoid overconfidence:
* Rather than stating "the mean is 20," a CI says, "we're 95% confident the mean is between 18 and 22," which is more informative and cautious.

Question 8: What is the concept of expected value in a probability distribution?

Answer: The expected value of a random variable is the long-run average of its outcomes, weighted by their probabilities. It tells the center or "balance point" of a probability distribution.

Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.
(Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

mean = 50
std_dev = 5
sample_size = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

computed_mean = np.mean(data)
computed_std = np.std(data)

print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std:.2f}")

plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
